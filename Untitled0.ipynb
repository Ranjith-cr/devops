```python
import gradio as gr
import fitz  # PyMuPDF
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "ibm-granite/granite-3.1-2b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")
model.eval()
```

```python
def generate_answer(prompt, max_new_tokens=512):
    full_prompt = f"Question:\n\n{prompt}\n\nAnswer:\n"
    inputs = tokenizer(full_prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return text.split("Answer:")[-1].strip()
```

```python
def studbud(topic, duration, pace, style, pdf_file_path):
    try:
        if not topic.strip():
            return "⚠️ Please enter a study topic."

        result = ""

        # Extract from PDF if uploaded
        if pdf_file_path:
            doc = fitz.open(pdf_file_path)
            extracted = "\n".join(page.get_text() for page in doc)
            MAX_LEN = 2000
            if len(extracted) > MAX_LEN:
                extracted = extracted[:MAX_LEN]

            insights_prompt = f"Explain the key points of this text:\n{extracted}"
            insights = generate_answer(insights_prompt)
            result += "**Insights from Uploaded Material:**\n" + insights + "\n\n"

        # Generate study plan
        prompt = (
            f"Create a {duration} study plan with a {pace} learning pace "
            f"focusing on {style} learning. Topic: {topic}"
        )
        plan = generate_answer(prompt)
        result += "**Your Personalized Study Plan:**\n" + plan

        return result
    except Exception as e:
        return f"⚠️ Error occurred: {e}"
```

```python
# Close old sessions (prevents port clash in Colab)
gr.close_all()

inputs = [
    gr.Textbox(
        label="Enter your study topic or question",
        lines=2,
        placeholder="e.g., Data Structures for Interviews"
    ),
    gr.Dropdown(["1 Week", "1 Month", "3 Months"], label="Study Duration", value="1 Month"),
    gr.Dropdown(["Slow", "Medium", "Fast"], label="Learning Pace", value="Medium"),
    gr.Dropdown(["Text-based", "Video-based", "Practice-heavy"], label="Preferred Study Style", value="Practice-heavy"),
    gr.File(label="Upload a study material (PDF)", file_types=[".pdf"], type="filepath")
]

demo = gr.Interface(
    fn=studbud,
    inputs=inputs,
    outputs=gr.Markdown(label="Plan"),
    title="📘 Studbud: AI Personalized Study Planner",
    description="Generates custom study plans (and insights from your materials) using Granite-3.1-2B-Instruct.",
    allow_flagging="never"
)

demo.launch(share=True, inline=True, debug=True)
```
